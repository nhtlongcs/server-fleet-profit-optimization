```python
# convert dict to df 
# {ts: {sg: {ls: value}}} -> {timesteps: [], sg: [], ls: [], value: []}
def convert_dict_to_df(data):
    timesteps = []
    sg = []
    ls = []
    value = []
    for ts, v1 in data.items():
        for sg_, v2 in v1.items():
            for ls_, value_ in v2.items():
                timesteps.append(ts)
                sg.append(sg_)
                ls.append(ls_)
                value.append(value_)
    return pd.DataFrame({'timesteps': timesteps, 'sg': sg, 'ls': ls, 'value': value})

df=convert_dict_to_df(data)
```
